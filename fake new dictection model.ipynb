### Modelling without Pre-processing Text data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.text,
    df.label_num,
    test_size=0.2,
    random_state=2022,
    stratify= df.label_num
)
```

```python
print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from  sklearn.neighbors import KNeighborsClassifier
```

```python
clf = Pipeline([
    ('vectorizer_trigrams', CountVectorizer(ngram_range = (1, 3))),
    ('KNN', (KNeighborsClassifier(n_neighbors=10, metric = 'euclidean'))) 
])
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

### Use text pre-processing to remove stop words, punctuations and apply lemmatization

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm") 
```

```python
def preprocessing(text):
    doc = nlp(text)
    filted_token = []
    for tokens in doc:
        if tokens.is_stop or tokens.is_punct:
            continue
        filted_token.append(tokens.lemma_)
    return " ".join(filted_token)
```

```python
df['preprocessed_text'] = df['text'].apply(preprocessing)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.preprocessed_text, 
    df.label_num,
    test_size=0.2,
    random_state=2022,
    stratify=df.label_num
)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = Pipeline([
    ('vectorizer_n_grams', CountVectorizer(ngram_range = (3, 3))),                       #using the ngram_range parameter 
    ('random_forest', (RandomForestClassifier()))         
])

clf.fit(X_train, y_train)


y_pred = clf.predict(X_test)


print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm


from matplotlib import pyplot as plt
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Prediction')
plt.ylabel('Truth')
```
